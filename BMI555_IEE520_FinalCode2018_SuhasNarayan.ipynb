```python
#converting the categorical to numerical values
def x5_to_numeric(x):

    
    if x=='A':
        return 1
    if x=='B':
        return 2
    if x=='C':
        return 3
    if x=='D':
        return 4
    if x=='E':
        return 5

def x13_to_numeric(z):
    if z=='alpha':
        return 1
    if z=='beta':
        return 2
    if z=='chi':
        return 3
    
def x64_to_numeric(c):
    if c=='Mp':
        return 1
    if c=='Ma':
        return 2
    if c=='Mk':
        return 3
    if c=='Mm':
        return 4
    
def x65_to_numeric(v):
    if v=='NPT':
        return 1
    if v=='NST':
        return 2
    if v=='PT':
        return 3
    if v=='ST':
        return 4
    
```

```python
from sklearn import metrics
from pandas_ml import ConfusionMatrix
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.neural_network import MLPClassifier
```

```python
#Oversampling to have the same number of both labels
overSamp=RandomOverSampler()
X_sam,y_sam=overSamp.fit_sample(X,y)
classpos = np.where(y_sam == 1)[0]
classneg = np.where(y_sam == -1)[0]
x_classpos = len(classpos)
x_classneg = len(classneg)
print("Post Sample balancing: 1:",x_classpos,"-1:",x_classneg)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
X_train, X_test, y_train, y_test = train_test_split(X_sam, y_sam, test_size=0.30, random_state=17)
```

```python
#performing the GridsearchCV
param_grid = { "criterion" : ["entropy"], "min_samples_leaf" : [5, 6, 10, 15], "min_samples_split" : [2, 4, 10], "n_estimators": [1,5,10,15],"max_depth":[2,5,9]}

from sklearn.model_selection import GridSearchCV, cross_val_score

rf = RandomForestClassifier(n_estimators=100, max_features='auto', oob_score=False, random_state=1,bootstrap=True,warm_start=False)

clf = GridSearchCV(estimator=rf, param_grid=param_grid)

clf.fit(X_train, y_train)

best_parameters=clf.best_params_
print(best_parameters)
```

```python
#Running the Random Forest Classifier
seedStart=97
modelnow=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='entropy',
            max_depth=9, max_features='auto', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=6, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=15,
            oob_score=False, random_state=seedStart,
            warm_start=False)


modelnow.fit(X_train,y_train)
yhat = modelnow.predict(X_train)
print (metrics.accuracy_score(y_train, yhat))
print (metrics.classification_report(y_train, yhat))
print (confusion_matrix(y_train, yhat))
```

```python
modelnow.fit(X_train,y_train)
y_pred = modelnow.predict(X_test)
print (metrics.accuracy_score(y_test, y_pred))
print (metrics.classification_report(y_test, y_pred))
print (confusion_matrix(y_test, y_pred))
```

```python
from sklearn import model_selection
seed=125
actuals=[]
probs=[]
hats=[]
kfold = model_selection.KFold(n_splits=10, shuffle=True, random_state=seed)
for train, test in kfold.split(X_sam, y_sam):
    x_train, x_test = X_sam[train], X_sam[test]
    Y_train, Y_test = y_sam[train], y_sam[test]
    modelnow.fit(x_train, Y_train)
    foldhats = (modelnow.predict(x_test))
    foldprobs = (modelnow.predict_proba(x_test))[:,1] 
    actuals = np.append(actuals, Y_test) 
    probs = np.append(probs, foldprobs)
    hats = np.append(hats, foldhats)

    
print (' Crossvalidation Accuracy = ', metrics.accuracy_score(actuals,hats))
print (metrics.classification_report(actuals, hats))
print (confusion_matrix(actuals, hats))
```
